## Q2. 각 일자의 요일을 표기하고 (‘Monday’ ~’Sunday’) ‘day_name’컬럼을 추가하고 이를 이용하여 각 요일별 이용 횟수의 총합을 데이터 프레임으로 출력하라

```python
df['대여일자'] = pd.to_datetime(df['대여일자'])
df['day_name'] = df['대여일자'].dt.day_name()
answer = df.day_name.value_counts().to_frame()
display(answer)
```

## Q3.각 요일별 가장 많이 이용한 대여소의 이용횟수와 대여소 번호를 데이터 프레임으로 출력하라 - 다시

```python
# groupby 뒤에 뭔가를 붙여줘야 하는점!! .sum() 이나 .size()
```

```python
df3 = df.groupby(['day_name', '대여소번호']).size().to_frame('size').sort_values(['day_name', 'size'], ascending=False).reset_index()
df3 = df3.drop_duplicates('day_name', keep = 'first').reset_index(drop = True)
df3
```
