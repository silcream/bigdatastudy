```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv', index_col = 0)
df.head()
```

## Q1. 인기 동영상 제작 횟수가 많은 채널 상위 10개명 출력 (날짜기준, 중복 포함)

```python
answer = list(df.loc[df.channelId.isin(df.channelId.value_counts().head(10).index)].channelTitle.unique())
answer
```

## Q2. dislikes 수가 like 수보다 높은 동영상 제작 채널 모두 출력

```python
answer2 = list(df.loc[df.likes<df.dislikes].channelTitle.unique())
answer2
```

## Q4. 일요일에 인기 있었던 영상들 중 가장 많은 영상 종류(categoryId)

```python
df.trending_date2
```

```python
## 시간정보가 있는 열을 datetime 자료형으로 변환
```

```python
df['trending_date2'] = pd.to_datetime(df['trending_date2'])
```

```python
df['weekday'] = df['trending_date2'].dt.weekday
```

```python
answer4 = df.loc[df.weekday == 0].categoryId.value_counts().index[0]
answer4
```

## Q5. 각 요일별 인기 영상들의 categoryId는 각각 몇개씩인지 하나의 데이터프레임으로 표현

```python
df
```

```python
group = df.groupby([df['trending_date2'].dt.day_name(), 'categoryId'], as_index = False).size() 
                                                                    #as_index 이 그룹을 인덱스로 지정할 것인가
group
```

```python
answer5 = group.pivot(index = 'categoryId', columns = 'trending_date2')
display(answer5)
```

## Q6. viewcount 대비 댓글 수 가 가장 높은 영상 확인 ( view_count 값이 0 인 경우 제외)

```python
df.info()
```

```python
df2 = df.loc[df.view_count != 0]
```

```python
df2['ratio'] = (df2['comment_count'] / df2['view_count']).dropna()
result = df2.sort_values(by = 'ratio', ascending = False).iloc[0].title
result
```

## Q8. like 대비 dislike 의 수가 가장 적은 영상은? 0인 경우 제외**

```python
df3 = df.loc[df.likes & df.dislikes != 0]
df3
```

```python
num = (df3['dislikes'] / df3['likes']).sort_values().index[0]
result = df.iloc[num].title ## df3  가 아닌 df 였다는점!!
```

```python
result
```
