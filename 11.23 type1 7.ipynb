## 대한민국 체력장 데이터

```python
import pandas as pd

df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/body/body.csv')
df.head()
```

전체데이터의 수축기혈압(최고) - 이완기혈압(최저)의 평균을 구하여라

```python
(df['수축기혈압(최고) : mmHg'] - df['이완기혈압(최저) : mmHg']).mean()
```

50~59세의 신장평균을 구하여라

```python
df50 = df[(df['측정나이'] >= 50) & (df['측정나이'] <= 59)]
df50
```

```python
df50['신장 : cm'].mean()
```

연령대 (20~29 : 20대 …) 별 인원수를 구하여라

```python
df['연령대'] = (df['측정나이']//10).astype(int)
df
```

```python
df.연령대.value_counts()
```

연령대 (20~29 : 20대 …) 별 등급의 숫자를 데이터 프레임으로 표현하라

```python
df.groupby(['연령대', '등급']).size().to_frame()
```

남성 중 A등급과 D등급의 체지방률 평균의 차이(큰 값에서 작은 값의 차)를 구하여라

```python
mdf = df[df['측정회원성별'] == 'M']
```

```python
(mdf[mdf['등급'] == 'D']['체지방율 : %'].mean()) - (mdf[mdf['등급'] == 'A']['체지방율 : %'].mean())
```

여성 중 A등급과 D등급의 체중의 평균의 차이(큰 값에서 작은 값의 차)를 구하여라

```python
fdf = df[df['측정회원성별'] == 'F']
```

```python
(fdf[fdf['등급'] == 'D']['체중 : kg'].mean()) - (fdf[fdf['등급'] == 'A']['체중 : kg'].mean())
```

bmi는 자신의 몸무게(kg)를 키의 제곱(m)으로 나눈값이다. 데이터의 bmi 를 구한 새로운 컬럼을 만들고 남성의 bmi 평균을 구하여라

```python
import numpy as np
```

```python
np.__all__
```

```python
df['bmi'] = df['체중 : kg']/np.square(df['신장 : cm']/100)
```

```python
df[df['측정회원성별'] == 'M'].bmi.mean()
```

bmi보다 체지방율이 높은 사람들의 체중평균을 구하여라

```python
df[df['bmi'] <  df['체지방율 : %']]['체중 : kg'].mean()
```

남성과 여성의 악력 평균의 차이를 구하여라

```python
mdf['악력D : kg'].mean() - fdf['악력D : kg'].mean()
```

남성과 여성의 교차윗몸일으키기 횟수의 평균의 차이를 구하여라

```python
abs(mdf['교차윗몸일으키기 : 회'].mean() - fdf['교차윗몸일으키기 : 회'].mean()) #절대값 - abs
```
